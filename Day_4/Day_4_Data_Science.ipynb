## Converting categorical into Numerical Data

```python
unique_values_of_Cabin = list(data["Cabin"].unique())
```

```python
unique_values_of_Cabin
```

```python
# Initializing the empty dictionary
ud = {}


for i in unique_values_of_Cabin:
    v = np.random.randint(0,9)
    if v in ud.values():
        x = np.random.randint(0,9)
        ud[i] = x
    else:
        ud[i] = v

        
# a = np.arange(0,10)
# b = np.array(list(ud.values()))
        
        
```

```python
list(ud.keys())
```

```python
# Mapping dictionary values to cabin manually

for i in range(len(data["Cabin"])):
    if data["Cabin"][i] in list(ud.keys()):
        data["Cabin"][i] = ud[data["Cabin"][i]]
```
