```python
x = np.arange(1,50,0.1)



p.figure(figsize= (15,7))

p.plot(x, np.sin(x), c = "red", linewidth = 5, label = "Sin")
p.plot(x, np.cos(x), c = "green", linewidth = 5, label = "Cos")
p.legend(["Sin", "Cos"])
p.xlabel("Values")
p.ylabel("Sin and Cos Values")
p.title("Sin VS Cos Graph")
p.show()
```

```python
x[:5]
```

```python
p.figure(figsize= (15,7))

p.scatter(x, np.sin(x), cmap = "summer")
p.scatter(x, np.cos(x), cmap = "summer")
p.legend(["Sin", "Cos"])
p.xlim(5,50)
p.ylim(-2,2)
p.xlabel("Values")
p.ylabel("Sin and Cos Values")
#p.grid()
p.style.use("seaborn")
p.title("Sin VS Cos Graph")
p.show()
```

```python
x = data["Doors"] * 100
```

```python
# Creatign seperate figures and axis
fig, ax = p.subplots(nrows = 2, ncols = 2, figsize = (15, 9))

# Accessign 1st plot and customizing it.
plot1 = ax[0][0].scatter(data["Colour"], data["Doors"])

ax[0][0].set(title = "DOORS VS COLOUR GRAPH", xlabel = "Colour", ylabel= " Doors")

# Accessign 2nd plot and customizing it.
plot2 = ax[0][1].plot(np.arange(1,50,0.1), np.sin(np.arange(1,50,0.1)), c = "Green")
plot2_1 = ax[0][1].plot(np.arange(1,50,0.1), np.cos(np.arange(1,50,0.1)), c = "Red")

ax[0][1].set(title = "Sin Graph", xlabel = "X Values", ylabel="Sin Values")
ax[0][1].legend(("Sin", "Cos"), loc = "upper right"   )


# Accessign 3rd plot and customizing it.
bar_plot = ax[1][0].bar(data["Make"], data["Doors"], color = data["Colour"])
ax[1][0].set(title = "Bar Graph", xlabel = "Make", ylabel="Doors")

# Accessign 4th plot and customizing it.
bar_plot = ax[1][1].hist(data["Doors"], bins = 5, color = "Red") 
ax[1][1].axhline(data["Doors"].mean())
ax[1][1].set(title = "Histogram", xlabel = "Doors", ylabel="Number of Doors")

#p.hist()
p.show()
```
